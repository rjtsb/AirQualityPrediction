```python
fig, ax = plt.subplots(figsize = (16, 10))
ax = s.barplot(x = "City", y = "SO2", data=df)
ax.set(ylabel = "SO2 level", title = "SO2 levels in city")
plt.show()
```

```python
# proportion by variables
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize = (10, 10), autopct = '%1.2f%%' , fontsize = 12)
    ax.set_title(variable + '\n', fontsize = 15)
    return n.round(dataframe_pie/df.shape[0]*100,2)
PropByVar(df,'City')
```

```python
#Heatmap

fig, ax = plt.subplots(figsize = (15, 10))
s.heatmap(df.corr(), ax = ax, annot = True)
```

```python
df.boxplot(column = "AQI", by = "AQI_Bucket")
```

```python
aqi = df["AQI"]
city = df["City"]
plt.subplots(figsize =(15, 5))
plt.plot(city, aqi, color = "g")
plt.ylabel('AQI level')
plt.xlabel('Cities')
plt.title('AQI level as per cities')
plt.show()
```
